## Matplotlib

### matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+. There is close resemble that of MATLAB. SciPy makes use of matplotlib as well. It is an excellent 2D and 3D graphics library for generating scientific figures.

### Some of the major Pros of Matplotlib are:

#### -  Generally easy to get started for simple plots
#### - Support for custom labels and texts
#### - Great control of every element in a figure
#### - High-quality output in many formats
#### - Very customizable in general

#### Matplotlib has pledged to not support Python 2 past 2020 by signing the Python 3 Statement. You can find more info at the official site https://matplotlib.org/

```python
# Let us import the main packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
```

```python
# Since we are using Jupyter notebooks we will use the following commabd. For other environments
# plt.show() at the end of all your plotting commands to have the figure pop up in another window.

%matplotlib inline
```

```python
# Let us create a simple example. We will need two np.arrays.

x = np.linspace(0, 5, 11)
y = x ** 2

print(x)
print(y)
```

```python
# Let us plot y over x with a specific colour.

plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title Here') # Specify X axis title
plt.ylabel('Y Axis Title Here') # Specify Y axis title
plt.title('String Title Here') # Specify general title
plt.show()
```

```python
# We can output multiple subplots on the same figure (similar with Matlab)

# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

## Matplotlib Object Oriented Method

Until now we saw some very basic concepts of Matplotlib. Matplotlib can be used for creating figure-objects and using Matplotlib's Object Oriented API to create figures (by calling methods or attributes).

```python
# We can create a empty figure object (an empty canvas) and specify exactly the wanted output figure.

# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title');
```

```python
# We can create a plot within another plot.

# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Plot 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Plot 2 Title');
```

```python
# As before we can create a figure with multiple subplots

# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, 'b')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Plot Y over X')

axes[1].plot(y, x, 'b')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
axes[1].set_title('Plot X over Y')

# Display the figure object    
fig;
```

```python
# Some times we have the problem of plots overlapping to each other. We can use fig.tight_layout() or plt.tight_layout() method,
# which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

plt.tight_layout()
```

```python
# We can specify the figure size and DPI.

fig = plt.figure(figsize=(8,4), dpi=100)

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title');
```

```python
# We can also save files. Go to the folder output_files and you can see the result. Matplotlib supports
# different types of files; including PNG, JPG, EPS, SVG, PGF and PDF. (It will give an error if the destination folder 
# doesn't exist already.)

fig.savefig("output_files/filename.png") # we could also specify the DPI by adding ", dpi="
```

```python
# We can plot multiple lines in the same figure and we can add a legend.

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, 'r', label="Red Line")
ax.plot(x, x**3, 'b', label="Blue Line")
ax.legend();
```

```python
# The legend function takes an optional keyword argument loc that can be used to specify where in the figure 
# the legend is to be drawn. The allowed values of loc are numerical codes for the various places the legend 
# can be drawn. See the documentation page at http://matplotlib.org/users/legend_guide.html#legend-location
# for details. Some of the most common loc values are:
# ax.legend(loc=1) # upper right corner
# ax.legend(loc=2) # upper left corner
# ax.legend(loc=3) # lower left corner
# ax.legend(loc=4) # lower right corner
# ax.legend(loc=0) # let matplotlib decide the optimal location


ax.legend(loc=4)
fig
```

```python
# We have already see that we can specify colours with MATLAB-like notation (e.g. 'b' for blue). We can also specify it
# in a different way. We can also specify how transparent the plots should be.

fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00");        # RGB hex code 
```

```python
# We can customize the plots a lot using the linewidth (or lw for short) and the linestyle (or ls for short).
# Let's see a few examples.

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
# We can specify the axis range of figures.

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```

## Pandas Visualization with Matplotlib

```python
# Let us import the data DataFrame that we have created in the Pandas notes, this DataFrame has the daily returns
# from Apple, Amazon, IBM, Netflix, and SPY.

data = pd.read_csv('files/data.csv', index_col=0, parse_dates=True)

data.head(3)
```

```python
# Let us plot everything with a very simple way to see what happens.

data.plot();
```

```python
# Plotting everything together doesn't give us a good image to understand anything.
# Let us plot only the daily returns of the SPY, and specify the figure size as well.

data['SPY'].plot(figsize=(12,8));
```

```python
# We can add much more details and even save the figure.

data['SPY'].plot(figsize=(12,8))
plt.ylabel('Daily Return')
plt.xlabel('Time')
plt.title('SPY')
plt.savefig('output_files/Daily Returns of SPY.png')
```

```python
# We can add the title in a more compact way.

data['SPY'].plot(figsize=(8,6), title='SPY');
```

```python
# We can also look only the daily returns of a specific time period as well as we can specify limits for the Y axis.

data['SPY'].plot(figsize=(8,6), title='SPY', xlim=['2015-01-01','2016-01-01'], ylim=[-0.03, 0.05]);
```

```python
# We specify colour and style as well.

data['SPY'].plot(figsize=(8,6), title='SPY', xlim=['2015-01-01','2016-01-01'], ls='--', c='g');
```

```python
# We can save the specified time period as well as the time series of the daily return

idx = data.loc['2015-01-01':'2015-03-01'].index
spy = data.loc['2015-01-01':'2015-03-01']['SPY']

print(idx)
print(spy.head())
```

```python
# Now we can get the indices and the time series and plot them.

fig, ax = plt.subplots()
ax.plot_date(idx, spy,'-')
plt.tight_layout()
plt.show()
```

```python
# There is a problem with the appearence of the dates, they overlap to each other.


fig, ax = plt.subplots()
ax.plot_date(idx, spy,'-')

fig.autofmt_xdate() # Auto fixes the overlap!
plt.tight_layout()
plt.show()
```

```python
# We could also add a grid to the figure.

fig, ax = plt.subplots()
ax.plot_date(idx, spy,'-')
ax.yaxis.grid(True)
ax.xaxis.grid(True)
fig.autofmt_xdate() # Auto fixes the overlap!
plt.tight_layout()
plt.show()
```

```python
# We could also change the format of the dates.

fig, ax = plt.subplots()
ax.plot_date(idx, spy,'-')
ax.yaxis.grid(True)
ax.xaxis.grid(True)

ax.xaxis.set_major_locator(dates.MonthLocator())  # Change the format of the dates
ax.xaxis.set_major_formatter(dates.DateFormatter('%b\n%Y'))

fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

```python
# We can add a minor axis as well (and present a different format of the dates).

fig, ax = plt.subplots()
ax.plot_date(idx, spy,'-')
ax.yaxis.grid(True)
ax.xaxis.grid(True)

ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%Y-%b'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator()) # Add the minor axis
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

### Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)
